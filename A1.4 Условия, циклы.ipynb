## А1.5.2.4 Задание 4 

Написать программу, которая на вход получает натуральное число n и выводит n-е число ряда Фибоначчи на экран. Для проверки используйте n = 10

Пояснение: последовательность Фибоначчи — это ряд чисел, в котором первые два элемента равны 1, а каждый следующий элемент равен сумме двух предыдущих: 1, 1, 2, 3, 5, 8, 13, 21, ...


```python
n = 10
fb1 = fb2 = 1

for i in range(2,n):
    if i != 1 or i != 2:
        fb1, fb2 = fb2, fb1+fb2
        print(fb2)
```

По сути запись fib1, fib2 = fib2, fib1 + fib2 равносильна другой записи:
temp = fib1
fib1 = fib2
fib2 = temp + fib2

Рассмотрим запись:
temp = fib1
fib1 = fib2
fib2 = temp + fib2

В программировании при обмене значений переменных обычно вводят третью переменную (temp), играющую роль буфера (ее иногда называют буферной переменной). В этой переменной (temp) сохраняют значение первой переменной (fib1), потом первой переменной (fib1) присваивают значение второй (fib2), a новое значение для второй переменной (fib2) берут из буфера (temp) и добавляют (+) значение еще не изменившейся второй переменной (fib2).

Запись fib1, fib2 = fib2, fib1 + fib2 по сути представляет такую вещь в Python как множественное присваивание
При выполнении fib1, fib2 = fib2, fib1 + fib2 интерпретатор Python сначала получает значения связанные с переменными fib2 и fib1 + fib2 (правая часть) и помещает их в кортеж (fib1, fib1+fib2) (почитайте, это структура данных такая). После этого он связывает каждый элемент кортежа в определенной позиции с переменной в той же позиции, но в кортеже слева (fib1, fib2).
